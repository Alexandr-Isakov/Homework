```python
import matplotlib.pyplot as plt
import numpy as np

ListX1 = []                                  # список координат по оси X у точек, по которым идёт интерполяция 
ListY1 = []                                  # список координат по оси Y у точек, по которым идёт интерполяция
ListX2 = []                                  # список всех точек по X на графике
ListY2 = []                                  # список всех значений по Y на графике
n = 13                                       # количество данных точек (по которым идёт интерполяция)
m = 1001                                     # количество точек, по которым будет строиться график
r = 1                                        # график строится на отрезке [-r ; r]

def f(x):                                    # функция, график которой пытаемся получить с помощью интерполяции
    return 1 / (1 + 25 * x ** 2)

for i in range(n):                           # создаются списки ListX1 и ListY1 
    ListX1.append(2 * r * i / (n - 1) - r)
    ListY1.append(f(ListX1[i]))

for i in range(m):
    ListX2.append(2 * r * i / (m - 1) - r)   # формируется массив точек по оси X, по которым строится график
    L = 0                                    # переменная значения f(x) в точке ListX2[i]
    for j in range(n):                       # формируется f(x) для точки ListX2[i]
        l = 1                                # переменная значения базисного полинома l с индексом j в точке ListX2[i]
        for k in range(n):                   # находится значение базисного полинома l с индексом j в точке ListX2[i]
            if k != j:
                l *= (ListX2[i] - ListX1[k]) / (ListX1[j] - ListX1[k])
        L += ListY1[j] * l
    ListY2.append(L)

plt.plot(ListX2, ListY2, '-', color='blue')  # получившийся график
plt.plot(ListX1, ListY1, 'o-', color='red')  # точки, по которым шла интерполяция 
plt.show()
```
